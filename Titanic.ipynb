```python
import pandas as pd

df = pd.read_csv("train.csv")
# data = df.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"])
data = df[["Survived", "Pclass", "Sex", "Age", "SibSp", "Parch", "Fare","Embarked"]]
data = data.dropna()

data.Sex = data.Sex.map({"female": 0, "male": 1})
data.Embarked = data.Embarked.map({"S": 0, "C": 1, "Q": 2})

# Standardize, PCA, Algo


data.head()
```

```python
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA


# new_X = scale(data.drop(columns=["Survived"]))

# pca = PCA(n_components=4)
# new_X = pca.fit_transform(new_X)

Y_train = data["Survived"].iloc[0:500]
X_train = data.drop(columns=["Survived"]).iloc[0:500]
# X_train = new_X[0:500, :]
Y_test = data["Survived"].iloc[500:]
X_test = data.drop(columns=["Survived"]).iloc[500:]
# X_test = new_X[500:, :]

```

```python
from sklearn.svm import SVC
from itertools import product

C_params = [0.2, 0.5, 1, 2, 5, 10, 25]
gamma_params = ["auto", "scale"]
kernel_params = ["linear", "rbf"]

for C, gamma, kernel in product(
    C_params, gamma_params, kernel_params): 
    cls = SVC(C=C, gamma=gamma, kernel=kernel)
    cls.fit(X_train, Y_train)
    score = cls.score(X_test, Y_test)
    print(f"C = {C}, gamma={gamma}, kernel={kernel} -> score = {score}")
```
